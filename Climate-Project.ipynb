Setelah memahami bahwa distribusi curah hujan memiliki kecenderungan tidak normal pada hari banjir, sekarang kita akan mengeksplorasi suhu udara sebagai faktor cuaca lainnya.
<br> Apakah suhu rata-rata berbeda antar bulan? Jika ya, bulan dengan suhu ekstrem bisa berdampak pada kondisi tanah, kelembapan udara, dan kerentanan pertanian terhadap banjir atau kekeringan.
<br> Maka dari itu dilakukan uji korelasi menggunakan One-Way ANOVA untuk menguji apakah rata-rata suhu (tavg) antar bulan berbeda secara statistik.

Hipotesis:
<br> H0 : Tidak ada perbedaan rata - rata suhu antar bulan. <br>
<br> H1 : Ada satu bulan yang memiliki rata - rata suhu berbeda dibanding bulan lain. <br>

```python
from scipy.stats import f_oneway

# Memfilter data suhu yang != 0
df_suhu = df_clean[['month', 'suhu_rata2']].dropna()

# Daftar suhu perbulan untuk uji ANOVA
bulan_list = ['January', 'February', 'March', 'April', 'May', 'June',
              'July', 'August', 'September', 'October', 'November', 'December']

# Pastikan urutan bulan
df_suhu['month'] = pd.Categorical(df_suhu['month'], categories=bulan_list, ordered=True)
df_suhu = df_suhu.sort_values('month')

# Siapkan grup suhu per bulan untuk ANOVA
groups = [df_suhu[df_suhu['month'] == bulan]['suhu_rata2'] for bulan in bulan_list]

# One-Way ANOVA
f_stat, p_value = f_oneway(*groups)
print(f"F-statistic: {f_stat:.2f}, p-value: {p_value:.4e}")
```

Dari hasil oneway ANOVA didapatkan nilai p-value : 2.3303e-122. p-value < 0.05 maka H0 ditolak dan H1 diterima yang berarti adalah ada setidaknya minimal satu bulan yang memiliki rata - rata suhu berbeda dibanding bulan lain.

Setelah mengetahui bahwa suhu berbeda secara signifikan antar bulan, maka dari itu penting dilakukan identifikasi wilayah mana saja yang memang secara statistik lebih sering terdampak banjir. Ini diperlukan agar petani bisa menghindari atau mengantisipasi risiko berdasarkan lokasi lahan.

Pada kasus ini digunakan analisis Chi-Square Test. 
<br> Hipotesis:<br>
<br> H0 : Tidak ada hubungan antara kejadian banjir dengan wilayah.<br>
<br> H1 : Ada hubungan antara kejadian banjir dengan wilayah.<br>

```python
from scipy.stats import chi2_contingency

contingency = pd.crosstab(df_clean['station_name'], df_clean['flood'])

# Uji Chi-Square
chi2, p_val, dof, expected = chi2_contingency(contingency)

# Tampilkan hasil
print(f"Hasil Chi-squared")
print(f"Chi-square statistic: {chi2:.4f}")
print(f"p-value: {p_val:.4e}")
print(f"Degrees of freedom: {dof}")
print(f"Expected: {expected}")
```

```python
# Menghitung Cramer's V

def hitung_cramers_v(chi2, n, min_dim):
    """
    Menghitung Cramer's V dari hasil Chi-squared
    Cramer's V = sqrt(chi2 / (n * (min_dim - 1)))
    dimana:
    chi2 = nilai chi-square
    n = ukuran sampel total
    min_dim = minimum dari jumlah baris atau kolom
    """
    return np.sqrt(chi2 / (n * (min_dim - 1)))

n = len(df)
min_dim = min(contingency.shape)
cramers_v = hitung_cramers_v(chi2, n, min_dim)

print(f"Cramer's V: {cramers_v:.3f}")
```

Dari hasil analisis, diperoleh nilai p-value 2.9382e-54 karena p-value < 0.05 maka H0 ditolak artinya Ada hubungan antara kejadian banjir dengan wilayah. 

Kekuatan hubungan (Cramer's V) adalah 0.198 karena nilai tersebut < 0.4 termasuk ke hubungan yang lemah maka dapat disimpulkan ada pengaruh kecil, tapi banyak faktor lain yang lebih dominan. Beberapa wilayah mungkin lebih rawan karena kontur tanah, infrastruktur drainase, atau curah hujan lokal.
